```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
```

```python
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df = iris_df.rename(columns={'petal length (cm)': 'petal length'})
iris_df
```

```python
x_axis = iris_df.index
y_axis = iris_df['petal length']
sns.scatterplot(x=x_axis, y=y_axis, hue=iris_df['target'])
```
